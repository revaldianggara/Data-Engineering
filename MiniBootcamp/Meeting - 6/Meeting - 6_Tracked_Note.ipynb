# Simulating Data Ingestion

```python
#import needed libraries
from sqlalchemy import create_engine
import pandas as pd
import os
import time
```

```python
#get password from environmnet var
username = "postgres"
password = "password"
server = "localhost"
database = "postgres"
port = "5432"

engine = create_engine(f'postgresql://{username}:{password}@{server}:{port}/{database}')
```

```python
# SIMLATE INGESTION OF FILE
local_data = pd.read_csv('./sales_generated.csv')
local_data.info()
```

```python
for index, row in local_data.head(100).iterrows():
    transform = pd.DataFrame(row.to_frame().T)
    # print(transform)
    transform.to_sql(f"sales_data", engine, if_exists='append', index=False)
    print("Row Inserted " + transform.date.astype(str) + ' ' + transform.name.astype(str) +  ' ' + transform.market_area.astype(str))
    time.sleep(3)
```

```python
import json

bootstrap_servers = ['localhost:29092']
topicName = 'sd.public.sales_data'
# Initialize consumer variable
consumer = KafkaConsumer (topicName , auto_offset_reset='earliest',
                          bootstrap_servers = bootstrap_servers, group_id='sales-data-dev-1')

# Read and print message from consumer
for msg in consumer:
    print(json.loads(msg.value))
```

# Manual Test (optional)

```python
# PRODUCER
# Import KafkaProducer from Kafka library
from kafka import KafkaProducer

# Define server with port
bootstrap_servers = ['localhost:29092']

# Define topic name where the message will publish
topicName = 'belajar_kafka'

# Initialize producer variable
producer = KafkaProducer(bootstrap_servers = bootstrap_servers)

# Publish text in defined topic
producer.send(topicName, b'Hello Kafka... Sedang Belajar 2')

# Print message
print("Message Sent")
```

```python
# CONSUMER
from kafka import KafkaConsumer

# Define server with port
bootstrap_servers = ['localhost:29092']

# Define topic name from where the message will recieve
topicName = 'source.public.sales_data'

# Initialize consumer variable
consumer = KafkaConsumer (topicName , auto_offset_reset='latest', bootstrap_servers = bootstrap_servers)

# Read and print message from consumer
for msg in consumer:
    print("Topic Name=%s,Message=%s"%(msg.topic,msg.value))
```
